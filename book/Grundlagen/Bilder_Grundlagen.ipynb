```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y = x**2


plt.plot(x,y)
plt. rc('axes', labelsize= 12) 
plt.xlabel(r"$x \rightarrow$")
plt.ylabel(r"$y = x^2 \rightarrow$")
#plt.title(r"Beispiel einer Funktion", fontdict = {'fontsize' : 15})
plt.savefig("Beispiel_Funktion.png", dpi = 100)
```

```python
x = np.linspace(0,10,100)
y = -3*x + x**2

plt.plot(x,y)
plt.rc('axes', labelsize= 12) 
plt.xlabel(r"$x \rightarrow$")
plt.ylabel(r"$y = f(x) = -3 \cdot x + x^2 \rightarrow$")
#plt.title(r"Beispiel für eine stetige eine Funktion", fontdict = {'fontsize' : 25})
plt.savefig("Beispiel_stetige_Funktion.png", dpi = 150)
```

```python
x = np.linspace(0,10,100)
y = x.copy()
for i in range(len(x)):
    if (x[i] < 5):
        y[i] = 0
    else: 
        y[i] = 1

plt.plot(x,y)
plt.rc('axes', labelsize= 12) 
plt.xlabel(r"$x \, \rightarrow$")
plt.ylabel(r"$y = \Theta(x-5) \, \rightarrow$")
#plt.title(r"Beispiel für eine nicht stetige Funktion", fontdict = {'fontsize' : 25})
plt.savefig("Beispiel_nicht_stetige_Funktion.png", dpi = 150)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y = x**2
dy = np.gradient(y)

plt.plot(x,dy)
plt. rc('axes', labelsize= 12) 
plt.xlabel(r"$x \rightarrow$")
plt.ylabel(r"$y^\prime = 2 \cdot x \rightarrow$")
#plt.title(r"Ableitung dieser Funktion", fontdict = {'fontsize' : 15})
plt.savefig("Beispiel_Ableitung.png", dpi = 100)
```
